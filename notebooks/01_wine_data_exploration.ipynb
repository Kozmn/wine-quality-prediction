```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

#Loading the dataset
dataset_white = pd.read_csv('data/winequality-red.csv', sep=';')

# Displaying the first few rows of the dataset
print(dataset_white.head())
print(dataset_white.describe())
print(f"shape: {dataset_white.shape}")
print(f"null count: {dataset_white.isnull().sum().sum()}")
print(f"columns: {dataset_white.columns}")

dataset_white.info()
```

```python
#Target = variable 'quality'
# Displaying the correlation matrix
correlation_matrix = dataset_white.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix of Wine Quality Dataset')
plt.show()

# Displaying the distribution of the target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='quality', data=dataset_white, palette='viridis')
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


```

```python
print("Correlation with wine quality")
quality_corr = dataset_white.corr()['quality'].sort_values(key=abs, ascending=False)
print(quality_corr)

print("\nDistribution of Quality (details)")
quality_counts = dataset_white['quality'].value_counts().sort_index()
quality_percent = (quality_counts / len(dataset_white) * 100).round(1)

for quality, count in quality_counts.items():
    print(f"Quality {quality}: {count} samples ({quality_percent[quality]}%)")
```

```python
plt.figure(figsize=(12, 10))
sns.boxplot(y='alcohol', x='quality', data=dataset_white, palette='viridis')
plt.title('Alcohol Content vs Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
plt.grid(True)
plt.show()
```
